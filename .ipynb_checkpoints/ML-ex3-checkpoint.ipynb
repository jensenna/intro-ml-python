Let's create a function that can perform the optimization (gradient descent)

```python
def gradientDescent(X, y, theta, alpha, iters):
    temp = np.matrix(np.zeros(theta.shape))
    parameters = int(theta.ravel().shape[1])
    cost = np.zeros(iters)
    
    for i in range(iters):
        error = (X * theta.T) - y
        
        for j in range(parameters):
            term = np.multiply(error, X[:,j])
            temp[0,j] = theta[0,j] - ((alpha / len(X)) * np.sum(term))
            
        theta = temp
        cost[i] = computeCost(X, y, theta)
        
    return theta, cost
```

Initializing learning rate (alpha) and number of iterations

```python
alpha = .01
iters = 1000
g, cost = gradientDescent(X, y, theta, alpha, iters)
g
```

Now let's compute the cost associated with these parameters:

```python
computeCost(X, y, g)
```
