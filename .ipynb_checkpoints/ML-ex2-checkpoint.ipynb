Let's work towards building logistic regression. To do so, we need the sigmoid function:

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
nums = np.arange(-10, 10, step = 1)

fig, ax = plt.subplots(figsize = (12,18))
ax.plot(nums, sigmoid(nums),'r')
```

```python
def computeCost(theta, X, y):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    ## need to convert to matrices so that it works with matrices as inputs
    first = np.multiply(-y, np.log(sigmoid(X * theta.T)))
    second = np.multiply((1 - y) , np.log(1 - sigmoid(X * theta.T)))
    ## np.multiply works with arrays and matrices
    return np.sum(first - second) / (len(X))
```

Unlike in the last problem, we are 'allowed' to use a built-in optimizer for this task, so we just need a function that calculates the gradient (the thing we are trying to optimize).

```python
def gradient(theta, X, y):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    
    parameters = int(theta.ravel().shape[1]) ##parameters will have same length as theta
    grad = np.zeros(parameters) ##gradient parameters will be set to filler values
    
    error = sigmoid(X * theta.T) - y  ##calc error
    for j in range(parameters):
        term = np.multiply(error, X[:,j]) ##multiply each error by each field
        grad[j] =  np.sum(term) / len(X)

    return grad
        
#     return theta, cost
```

Finally, we can use SciPy's optimize namespace to find the minimized cost.

```python
import scipy.optimize as opt
result = opt.fmin_tnc(func=computeCost, x0=theta, fprime=gradient, args=(X, y))
result
```

```python
max_degree = 6

def map_feature(test_1, test_2, M):
    """Python version of mapFeature.m."""
    # Add a column of ones to the data matrix that allows us to treat the intercept parameter as a feature.
    X_T = np.ones(M) 
    
    for i in range(1, max_degree+1):
        for j in range(i+1):
            X_T = np.vstack((X_T, test_1**(i-j) * test_2**j))
            
    return X_T.T
```

Now that we have our polynomials, we need to modify the cost function in order to reflect the regularization.

```python
# note, do not regularize the intercept!
def computeRegCost(theta, X, y, learningRate):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    ## need to convert to matrices so that it works with matrices as inputs
    first = np.multiply(-y, np.log(sigmoid(X * theta.T)))
    second = np.multiply((1 - y) , np.log(1 - sigmoid(X * theta.T)))
    reg = learningRate / (2 * len(X)) * np.sum(np.power(theta[:,1:theta.shape[1]],2))
    return np.sum(first - second) / (len(X)) + reg
```

We also need to update the gradient descent to incorporate regularization

```python
def gradientReg(theta, X, y, learningRate):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    
    parameters = int(theta.ravel().shape[1]) ##parameters will have same length as theta
    grad = np.zeros(parameters) ##gradient parameters will be set to filler values
    
    error = sigmoid(X * theta.T) - y  ##calc error
    for j in range(parameters):
        term = np.multiply(error, X[:,j]) ##multiply each error by each field
        
        if(j == 0):
            grad[j] =  np.sum(term) / len(X)
        else:
            grad[j] =  (np.sum(term) / len(X)) + ((learningRate / len(X)) * theta[:,j])
    return grad
        
#     return theta, cost
```
